```python

#W7A8
import ast
n = ast.literal_eval(input())
k = ast.literal_eval(input())
def find(n, k):
    lst = []
    for i in n:
        if i[0] <= k <= i[1]:
            lst.append(i[1]-i[0]+1)
    if len(lst) == 0:
        return -1
    else:
        return min(lst)
lst = []
for i in k:
    lst.append(find(n,i))
print(lst)
```
