```python
#W8A3
try:
    def tong_bp(n):
        n = str(n)
        s = 0
        for i in n:
            s += int(i)**2
        return s
    kq = []
    n = int(input())
    if n == 1:
        print("YES")
    while n != 1 and n not in kq:
        n = tong_bp(n)
        kq.append(n)
    if 1 in kq:
        print("YES")
    else:
        print("NO")
except ValueError:
    print("NO")
```

```python
#W10A10
n = list(map(int,input().split()))
k = n[0]
mang = n[1:len(n)]
try:
    if len(mang) != k:
        raise
    dic1 = {}
    for i in mang:
        if i not in dic1:
            dic1[i] = 1
        else:
            dic1[i] += 1 
    for x,y in dic1.items():
        if y > 1:
            raise
    s=0
    for i in mang:
        s+=i
    print(s)
except:
    print("Mang khong hop le")
```

```python
#W8A11
import re
def check_regex(s):
    try:
        re.compile(s)
        return True
    except re.error:
        return False
k = int(input())
lst = []
for i in range(k):
    lst.append(input())
for i in lst:
    print(check_regex(i))
```
