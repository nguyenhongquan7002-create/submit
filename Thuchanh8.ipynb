```python
#W8A1
class Cylinder:
    def __init__(self, r,h):
        self.r = r
        self.h= h
    def Sbemat(self):
        pi = 3.14
        S = 2*pi*self.r**2 + 2*pi*self.r*self.h
        return S
    def Thetich(self):
        pi = 3.14
        V = self.h*pi*self.r**2
        return V
n = list(map(int,input().split()))
S = Cylinder(n[0],n[1]).Sbemat()
V = Cylinder(n[0],n[1]).Thetich()
print(f'{S:.2f} {V:.2f}')
```

```python
#W8A2
class Date:
    def __init__(self,dd,mm,yy):
        self.dd = dd
        self.mm = mm
        self.yy = yy
    def is_leap(self):
        return self.yy%400==0 or (self.yy%4 == 0 and self.yy%100 != 0)
    def max_day_in_month(self):
        max_day = [0,31,28,31,30,31,30,31,31,30,31,30,31]
        if self.is_leap():
            max_day[2] = 29
        if 1<=self.mm <= 12:
            return max_day[self.mm]
        else:
            return 0
    def is_valid(self):
        if self.yy < 0 or self.mm < 1 or self.mm > 12 or self.dd < 1 or self.dd > self.max_day_in_month():
            return False
        return True
    
    def next_day(self):
        if self.is_valid() == False:
            return "INVALID"
        self.dd +=1
        if self.dd > self.max_day_in_month():
            self.dd = 1
            self.mm += 1
            if self.mm > 12:
                self.mm = 1
                self.yy +=1
        return f"{self.dd:02d}/{self.mm:02d}/{self.yy:04d}"
n = input()
n = n.replace('/',' ')
n = n.split()
n = list(map(int,n))
print(Date(n[0],n[1],n[2]).next_day())
```

```python
#W8A3
import math
class toa_do:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def ktra(self):
        if self.x == self.y == 0:
            return "Co la goc toa do"
        elif self.x == 0:
            return f"Co nam tren truc tung va khoang cach la {abs(self.y)}"
        elif self.y == 0:
            return f"Co nam tren truc hoanh va khoang cach la {abs(self.x)}"
        else:
            return f"Khoang cach la {(self.x**2 + self.y**2)**0.5:.2f}"
x,y = list(map(int,input().split()))
print(toa_do(x,y).ktra())
```

```python
#W8A4
class Calculator:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def add(self):
        return self.a + self.b
    def set_numbers(self,new_a,new_b):
        self.a = new_a
        self.b = new_b
    def subtract(self):
        return self.a - self.b
    def product(self):
        return self.a * self.b
    def divide(self):
        if self.b != 0:
            return self.a / self.b
        else:
            return "Lỗi phép tính"
    def power(self):
        return self.a ** self.b
    def mod(self):
        if self.b != 0:
            return self.a%self.b
        else:
            return "Lỗi phép tính"
def show_menu():
    print("\n--- MENU MÁY TÍNH ---")
    print("1. Cộng (add)")
    print("2. Trừ (subtract)")
    print("3. Nhân (product)")
    print("4. Chia (divide)")
    print("5. Hàm mũ (power)")
    print("6. Lấy dư (mod)")
    print("7. Cập nhật 2 số mới (set_numbers)")
    print("----------------------")
a, b = list(map(float, input().split()))
my_calc = Calculator(a, b) 
while True:
    show_menu()
    choice = input("Chọn phép tính: ")
    if choice == '1':
        print(f"Kết quả: {my_calc.add()}")
    elif choice == '2':
        print(f"Kết quả: {my_calc.subtract()}")
    elif choice == '3':
        print(f"Kết quả: {my_calc.product()}")
    elif choice == '4':
        print(f"Kết quả: {my_calc.divide()}")
    elif choice == '5':
        print(f"Kết quả: {my_calc.power()}")
    elif choice == '6':
        print(f"Kết quả: {my_calc.mod()}")
    elif choice == '7':
        n1, n2 = list(map(int,input().split()))
        my_calc.set_numbers(n1, n2) 
        print("Đã cập nhật số mới thành công!")
    else:
        print("Lựa chọn không hợp lệ.")

    quit_answer = input("Bạn có muốn thoát?")
    if quit_answer.lower() == "yes":
        break
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.lst1 = []
    def add(self, name, price):
        dic1 = {name:price}
        self.lst1.append(dic1)
        print(f"Đã thêm {name} vào giỏ hàng.")
    def remove(self, name):
        s = 0
        for i in self.lst1:
            if name in i:
                self.lst1.remove(i)
                print(f"Đã xóa {name} khỏi giỏ hàng")
                s = 1
        if s == 0:
            print(f"{name} không tồn tại trong giỏ hàng")
    def check(self):
        if len(self.lst1) == 0:
            print("Giỏ hàng rỗng")
        else:
            print("Giỏ hàng không rỗng")
    def sum(self):
        s = 0 
        for i in self.lst1:
            for x,y in i.items():
                s += y 
        print(f"Tổng tiền giỏ hàng là: {s}")
    
    def display(self):
        print(self.lst1)
    def remove_all(self):
        self.lst1 = []
        print("Đã xóa toàn bộ giỏ hàng")
cart = ShoppingCart()

while True:
    print("\n========= MENU =========")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm")
    print("4. Kiểm tra giỏ hàng rỗng")
    print("5. Tính tổng tiền")
    print("6. Hiển thị giỏ hàng")
    print("7. Xóa sạch giỏ hàng")
    
    choice = input("Nhập lựa chọn của bạn (2-7): ")
    if choice == '2':
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá sản phẩm: "))
        cart.add(name, price)
    elif choice == '3':
        name = input("Nhập sản phẩm muốn xóa: ")
        cart.remove(name)
    elif choice == '4':
        cart.check()
    elif choice == '5':
        cart.sum()
    elif choice == '6':
        cart.display()
    elif choice == '7':
        cart.remove_all()
    else:
        print("Lựa chọn không hợp lệ")
    question = input("Có muốn thoát khỏi chương trình không? ")
    if question == 'yes':
        break
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h, k):
        self.w = w
        self.h = h
        self.k = k 
    def area(self):
        print(f'{self.w * self.h * (self.k**2):.2f}', end = ' ')
    def perimeter(self):
        print(f'{(self.w+self.h)*2*self.k:.2f}')
w, h , k = list(map(float,input().split()))
Rectangle(w,h,k).area(), Rectangle(w,h,k).perimeter()
```

```python
#W8A7
import math
try:
    class Fraction:
        def __init__(self, a, b, c, d):
            self.a = a
            self.b = b
            self.c = c 
            self.d = d
            
        def add(self): return self.a*self.d + self.b*self.c, self.b*self.d
        def sub(self): return self.a*self.d - self.b*self.c, self.b*self.d
        def mul(self): return self.a*self.c, self.b*self.d
        def div(self): return self.a*self.d, self.b*self.c

    def rut_gon(tu, mau):
        if mau == 0: return "Lỗi phép tính"
        uoc = math.gcd(tu, mau)
        tu //= uoc
        mau //= uoc
        if mau < 0:
            tu, mau = -tu, -mau
        return f'{tu}/{mau}'

    a, b, op, c, d = input().split()
    f = Fraction(int(a), int(b), int(c), int(d))

    if op == '+': res = f.add()
    elif op == '-': res = f.sub()
    elif op == '*': res = f.mul()
    elif op == '/': res = f.div()

    print(rut_gon(res[0], res[1]))

except Exception:
    print("Lỗi phép tính")
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)
    def deposit(self, val):
        self.balance += val
    def withdram(self,val):
        if self.balance >= val:
            self.balance -= val
    def __str__(self):
        return f"{self.owner} : {self.balance} "

owner, balance = list(map(str,input().split()))
bank = BankAccount(owner, balance)
n = int(input())
for i in range(n):
    m = list(map(str,input().split()))
    op = m[0]
    x = float(m[1])
    if op == 'DEPOSIT':
        bank.deposit(x)
    elif op == 'WITHDRAW':
        bank.withdram(x)
print(f'{bank.balance:.2f}')
```

```python
#W8A9
class Student:
    def __init__(self,name):
        self.name = name
        self.score = []
    def add_score(self, scores):
        self.score.append(scores)
    def avg(self):
        return sum(self.score) / len(self.score)
    def rank(self):
        avgs = self.avg()
        if avgs >= 8:
            return "Excellent"
        elif 6.5 <= avgs < 8:
            return "Good"
        elif 5 <= avgs < 6.5:
            return "Average"
        else:
            return "Poor"
    def __str__(self):
        return f'{self.name} {self.avg():.2f} {self.rank()}'
name = input()
n = int(input())
hs = Student(name)
for i in range(n):
    mon, diem = list(map(str,input().split()))
    diem = float(diem)
    hs.add_score(diem)
print(hs)
```

```python
#W8A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.libra = []
    def add(self, book):
        self.libra.append(book)
    def count_by_author(self, name):
        count = 0 
        for book in self.libra:
            if name == book.author:
                count += 1 
        return count
    def count_by_year(self, year):
        count = 0
        for book in self.libra:
            if year == book.year:
                count += 1 
        return count
li = Library()
n = int(input())
for i in range(n):
    new = list(map(str,input().split(maxsplit=1)))
    ki_tu = new[0]
    if ki_tu == 'ADD':
        m = new[1].split(';')
        title = m[0]
        author = m[1]
        year = m[2]
        li.add(Book(title,author,year))
    elif ki_tu == "COUNT":
        print(li.count_by_author(new[1]))
    elif ki_tu == "COUNTYEAR":
        print(li.count_by_year(new[1]))
    
        
```

```python
#W8A11
import math     
x1,y1,z1 = list(map(float,input().split()))
x2,y2,z2 = list(map(float,input().split()))
x3,y3,z3 = list(map(float,input().split()))
x4,y4,z4 = list(map(float,input().split()))
ab = [x2-x1,y2-y1,z2-z1]
ac = [x3-x1,y3-y1,z3-z1]
tich_co_huong1 = [ab[1]*ac[2]-ac[1]*ab[2], ab[2]*ac[0]-ab[0]*ac[2], ab[0]*ac[1]-ab[1]*ac[0]]
bc = [x3-x2,y3-y2,z3-z2]
bd = [x4-x2,y4-y2,z4-z2]
tich_co_huong2 = [bc[1]*bd[2]-bd[1]*bc[2], bc[2]*bd[0]-bc[0]*bd[2], bc[0]*bd[1]-bc[1]*bd[0]]
tich_vo_huong = tich_co_huong1[0]*tich_co_huong2[0] + tich_co_huong1[1]*tich_co_huong2[1] + tich_co_huong1[2]*tich_co_huong2[2]
tich_do_dai = ((tich_co_huong1[0]**2 + tich_co_huong1[1]**2 + tich_co_huong1[2]**2)*(tich_co_huong2[0]**2 + tich_co_huong2[1]**2 + tich_co_huong2[2]**2))**0.5
cos = tich_vo_huong / tich_do_dai
phi =  math.acos(abs(cos))
phi = math.degrees(phi)
print(f'{phi:.2f}')
```

```python
#W8A12
class Complex:
    def __init__(self,re,im):
        self.re = re
        self.im = im
    def product(self,other):
        Re = self.re*other.re - self.im*other.im
        Im = self.re*other.im + self.im*other.re 
        return f'{Re} {Im}'
re,im = list(map(int,input().split()))
re1,im1 = list(map(int,input().split()))
comp = Complex(re,im)
other_comp = Complex(re1,im1)
comp.product(other_comp)
print(comp.product(other_comp))
```

```python
#W8A13
class Triangle:
    def __init__(self, egde1, egde2, egde3):
        self.egde1 = egde1
        self.egde2 = egde2
        self.egde3 = egde3
    def check(self):
        if self.egde3 + self.egde2 <= self.egde1 or self.egde1 + self.egde2 <= self.egde3 or self.egde3 + self.egde1 <= self.egde2:
            return False
        else:
            return True
    def dien_tich(self):
        nua_chu_vi = (self.egde1+self.egde2+self.egde3)/2
        dien_tich_binh = nua_chu_vi*(nua_chu_vi-self.egde3)*(nua_chu_vi-self.egde2)*(nua_chu_vi-self.egde1)
        res = dien_tich_binh**0.5
        if res == int(res):
            return str(int(res))
        else:
            return f'{res:.2f}' 
    def __str__(self):
        if self.check() == True:
            return self.dien_tich()
        else:
            return "invalid"
a,b,c = list(map(int,input().split()))
tam_giac = Triangle(a,b,c)
print(tam_giac)    
```
