```python
#W5A1
def Max_Of_Two(a,b):
    if a > b:
        return a
    elif a < b:
        return b
a,b = list(map(int,input().split()))
print(Max_Of_Two(a,b))
```

```python
#W5A2
def swap(a,b):
    a, b = b, a 
    return a, b 
a,b = list(map(int,input().split()))
print(*swap(a,b)) 
```

```python
#W5A3
def prime(n):
    s = 0
    for i in range(1, n+1):
        if n%i == 0:
            s += 1
    if s == 2:
        return True
    else:
        return False
print(prime(int(input())))
```

```python
#W5A4
def per(n):
    s = 0 
    for i in range(1,n):
        if n%i == 0:
            s+=i
    if s == n:
        return True
    else:
        return False
print(per(int(input())))
    
```

```python
#W5A5
def find(a, b):
    s=0
    for i in a:
        s+=1
        if i == b:
            return a[s-1]
    if b not in a:
        return -1
a = list(map(int,input().split()))
b = int(input())
print(find(a,b))
```

```python
#W5A6
def factorial(n):
    s = 1
    for i in range(1,n+1):
        s*=i
    return s
print(factorial(int(input())))
```

```python
#W5A7
def operat(a,b,c):
    if b == '+':
        return float(a) + float(c)
    elif b == "-":
        return float(a) - float(c)
    elif b == "*":
        return round((float(a) * float(c)),2)
    elif b == "/":
        return round((float(a) / float(c),2))
a,b,c = list(map(str,input().split()))
print(operat(a,b,c))
```

```python
#W5A9
def sum(n):
    s = 0
    for i in n:
        s += int(i)
    return s
print(sum(input()))
```

```python
#W5A10
def is_isomorphic_simple(a, b):
    n = len(a)
    for i in range(n):
        for j in range(i):
            if a[i] == a[j] and b[i] != b[j]:
                return False
            if a[i] != a[j] and b[i] == b[j]:
                return False
    return True
a, b = list(map(str,input().split()))
print(is_isomorphic_simple(a,b))
```
