```python
#W6A3
n = list(map(int,input().split()))
k = int(input())
s = []
d = []
if k <= len(n):
    for i in range(k,0,-1):
        s.append(n[len(n)-i])
    for i in range(k+1):
        d.append(n[i])
    print(tuple(s + d))
else:
    print(tuple(n))
```

```python
#W6A9
import ast
n = ast.literal_eval(input())
m = ast.literal_eval(input())
kq = {}
for x, y in n.items():
    if x not in kq:
        kq[x] = y 
    elif x in kq:
        kq[x] += y
for z,t in m.items():
    if z not in kq:
        kq[z] = t
    else:
        kq[z] += t
print(dict(sorted(kq.items())))
```
