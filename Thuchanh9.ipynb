```python
#W9A1
n = list(map(int,input().split()))
k = list(map(int,input().split()))
m = n + k
def arrange(m):
    for i in range(len(m)):
        for j in range((len(m)-i-1)):
            if m[j] > m[j+1]:
                m[j] , m[j+1] = m[j+1] , m[j]
    return m
print(*arrange(m))
```

```python
#W9A4
def solve():
    a = input()
    b = input()
    n, m = len(a), len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True
    for i in range(1, n + 1):
        for j in range(m + 1):
            if dp[i-1][j]:
                if a[i-1].islower():
                    dp[i][j] = True
            
            if j > 0 and dp[i-1][j-1]:
                if a[i-1].upper() == b[j-1]:
                    dp[i][j] = True
    print("YES" if dp[n][m] else "NO")
q = int(input())
for _ in range(q):
    solve()
```

```python
#W9A6
def hoanvi(n, k):
    a = -1
    for i in range(n-1,0,-1):
        if k[i] > k[i-1]:
            a = i-1
            break
    if a == -1:
        k.reverse()
        return k 
    for i in range(n-1,a,-1):
        if k[i] > k[a]:
            k[a], k[i] = k[i], k[a]
            break 
    return k[:a+1] + k[a+1:][::-1]
n = int(input())
k = list(map(int,input().split()))
print(hoanvi(n,k))
```

```python
#W9A7
def CanPlaceFlowers(flowerbed, k):
    n = len(flowerbed)
    s = 0
    if n == 1 and flowerbed[0] == 0 and k == 1:
        return True
    for i in range(n):
        if flowerbed[i] == 0 and n > 1 :
            if i == 0 and flowerbed[1] == 0:
                flowerbed[0] = 1 
                s+=1
            elif i == n-1 and flowerbed[n-2] == 0:
                flowerbed[n-1] = 1
                s+=1
            elif flowerbed[i-1] == flowerbed[i+1] == 0:
                flowerbed[i] = 1
                s+=1
    if s >= k:
        return True
    return False
flower = list(map(int,input().split()))
k = int(input())
print(CanPlaceFlowers(flower,k))
```

```python
#W9A8
def is_path_crossing(moves):
    x, y = 0,0
    lst1 = [[0,0]]
    for i in moves:
        if i == 'R':
            x += 1
        elif i == 'L':
            x -= 1
        elif i == 'U':
            y += 1
        elif i == 'D':
            y -= 1 
        toa_do_moi = [x,y]
        if toa_do_moi in lst1:
            return True
        else:
            lst1.append(toa_do_moi)
    return False
move = input()
print(is_path_crossing(move))
```

```python
#W9A9
def Media(nums):
    a,b,c,d,e = nums
    if a > b: 
        a,b = b,a 
    if c > d:
        c,d = d,c
    if a < c:
        a = e
        if a > b:
            a,b = b,a 
    else:
        c = e 
        if c > d:
            c,d = d,c 
    if a < c:
        if b < c:
            return b 
        else:
            return c 
    else:
        if a > d:
            return d
        else:
            return a
num = list(map(int,input().split()))
print(Media(num))
```

```python
#W9A10
import bisect

def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    
    max_radius = 0
    n = len(heaters)
    
    for house in houses:
        idx = bisect.bisect_left(heaters, house)
        dist_left = float('inf')
        dist_right = float('inf')
        if idx < n:
            dist_right = heaters[idx] - house
        if idx > 0:
            dist_left = house - heaters[idx - 1]
        shortest_to_this_house = min(dist_left, dist_right)
        max_radius = max(max_radius, shortest_to_this_house)
        
    return max_radius
if __name__ == "__main__":
    house = list(map(int,input().split()))
    heat = list(map(int,input().split()))
    print(findRadius(house,heat))
```
