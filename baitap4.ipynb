```python
def reverse(n):
    s = 0
    while n >= 1:
        m = n % 10
        s = s * 10 + m
        n = n // 10
    return s
n = int(input())
b = 0
while n != reverse(n):
    n = n + reverse(n)
    b += 1
print(b, n)
```
