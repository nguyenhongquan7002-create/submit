```python
#W11A4
def anagrams(path1, path2):

    
    def count(name):
        name = name.lower()
        kq = ''
        for i in name:
            if "a" <= i <= "z":
                kq += i
        kq = sorted(kq)
        dic = {}
        for i in kq:
            if i not in dic:
                dic[i] = 1
            else:
                dic[i] += 1
        return dic
    with open(path1, 'r') as f1:
        s = f1.read()
        dic1 = count(s)
    with open(path2, 'r') as f2:
        b = f2.read()
        dic2 = count(b)
    if dic1 == dic2:
        return True
    else:
        return False


n = input()
m = input()
print(anagrams(n,m))
    
```

```python
#W11A11
def max_min(path):
    try:
        with open(path) as f:
            s = f.readlines()
            lst = []
            for i in s:
                i = i.strip().replace('/',' ').split()
                i = list(map(int,i))
                for j in i:
                    lst.append(j)
            print(max(lst),min(lst))
    except:
        print("Mission failed")

        
pa = input()
max_min(pa)
```

```python
#W11A12
def maximumProduct(path):
    with open(path) as f:
        a = f.readline()
        s = f.readline()
        s = s.split()
        lst = []
        for i in s:
            lst.append(int(i))
        lst = sorted(lst)
        s1 = lst[0] * lst[1] * lst[-1]
        s2 = lst[-1] * lst[-2] * lst[-3]
        if s1 > s2:
            return s1
        else:
            return s2
        
        
pa = input()
print(maximumProduct(pa))
```

```python
#W11A13
def moveZeroes(fileName):
    with open(fileName) as f:
        a = f.readline()
        s = f.readline()
        s = s.split()
        lst1 = []
        lst0 = []
        for i in s:
            if int(i) == 0:
                lst0.append(0)
            else:
                lst1.append(int(i))
        return lst1 + lst0
    
    
file = input()
print(moveZeroes(file))
```

```python
#W11A14
def averageTime(path):
    with open(path) as f:
        s = f.readlines()
        lst = []
        for i in s[1:]:
            i = i.strip().split(',')
            lst.append(float(i[1]))
        tb = sum(lst) / len(lst)
        return f'{tb:.2f}'

        
pa = input()
print(averageTime(pa))
```

```python
#W11A15
def findMovies(path):
    with open(path) as f:
        s = f.readlines()
        lst = []
        for i in s[1:]:
            i = i.strip().split(',')
            if float(i[3]) >= 8:
                lst.append(i[1])
        if len(lst) == 0:
            return "boring"
        else:
            return lst
        
        
pa = input()
print(findMovies(pa))
```

```python
#W11A16
def chuvi(path):
    with open(path) as f:
        s = f.readlines()
        lst = []
        for i in range(0,len(s),3):
            s[i] = s[i].strip().split(':')
            if s[i][1].strip() =='SQUARE':
                so_lieu = s[i+1].strip().split(':')
                cv = round(4*float(so_lieu[1]),2)
                lst.append(cv)
            elif s[i][1].strip() == 'RECTANGLE':
                so_lieu = s[i+1].strip().replace(':',' ').split()
                cv = float(so_lieu[1]) + float(so_lieu[3])
                cv = round(cv*2,2)
                lst.append(cv)
            elif s[i][1].strip() == 'CIRCLE':
                so_lieu = s[i+1].strip().split(':')
                cv = round(2 * 3.14 * float(so_lieu[1]),2)
                lst.append(cv)
        return lst
    
    
pa = input()
lst = chuvi(pa)
for i in lst:
    print(f'{i:.2f}')
```

```python
#W11A17
def power(arr, i): 
    tem = 1 
    for j in range(0,len(arr)):
        if j == i:
            continue
        else: 
            tem *= arr[j]
    return tem


with open("input.txt", "r") as file: 
    n = file.readline()
    s = file.readline()
    s = s.split() # always have this step 
    lst = [int(n) for n in s]
    res = []
    for i in range(0,len(lst)): 
        res.append(power(lst,i))

        
    print(res)
```

```python
#W11A20
import math

try:
    with open("input.txt", "r") as file:
        s = file.read().split()
        M = int(s[0])
        raw = [int(x) for x in s[1:]]
        img = []
        for i in range(M):
            img.append(raw[i*M : (i+1)*M])
        res = [[0] * M for _ in range(M)]
        for r in range(M):
            for c in range(M):
                sum = 0
                count = 0
                for i in range(r - 1, r + 2):
                    for j in range(c - 1, c + 2):
                        if 0 <= i < M and 0 <= j < M:
                            sum += img[i][j]
                            count += 1
                res[r][c] = sum // count
        for row in res:
            print(*(row))
            
except:
    print("Mission failed")
```
